```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Step 1: Prepare the data
target_column = 'Lunch Calories'  # Correctly define the target column
X = a  # Features
y = merged_df[target_column]  # Target regression values

# Normalize the target values
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))  # Reshape for scaling

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_scaled, test_size=0.2, random_state=42)

# Step 2: Define the custom RMSRE loss function
def rmsre_loss(y_true, y_pred):
    # Add a small epsilon to avoid division by zero
    epsilon = tf.keras.backend.epsilon()
    relative_error = (y_true - y_pred) / (y_true + epsilon)
    squared_error = tf.square(relative_error)
    mean_squared_error = tf.reduce_mean(squared_error)
    rmsre = tf.sqrt(mean_squared_error)
    return rmsre

# Step 3: Build the ANN model
model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),  # Input layer
    layers.Dense(64, activation='relu'),      # Hidden layer with 64 units
    layers.Dense(32, activation='relu'),      # Hidden layer with 32 units
    layers.Dense(16, activation='relu'),      # Hidden layer with 16 units
    layers.Dense(1)                           # Output layer with 1 regression output
])

# Step 4: Compile the model
model.compile(optimizer='adam', loss=rmsre_loss)  # Use the custom RMSRE loss function

# Step 5: Train the model
history = model.fit(X_train, y_train, epochs=500, batch_size=32, validation_split=0.2)

# Step 6: Evaluate the model
test_loss = model.evaluate(X_test, y_test)
print(f'Test Loss (RMSRE): {test_loss}')

# Step 7: Make predictions
predictions = model.predict(X_test)

# Rescale predictions to the original scale
predictions_rescaled = scaler_y.inverse_transform(predictions)
print(f'Rescaled Predictions: {predictions_rescaled[:5]}')  # Display the first 5 predictions
```

```python
import keras_tuner as kt
from tensorflow.keras.optimizers import Adam

# Define the model as a hypermodel
def build_model(hp):
    model = keras.Sequential()
    model.add(layers.Input(shape=(X_train.shape[1],)))

    # Tune the number of layers and units in each layer
    for i in range(hp.Int("num_layers", 2, 5)):  # 2 to 5 hidden layers
        model.add(
            layers.Dense(
                units=hp.Int(f"units_{i}", min_value=16, max_value=128, step=16),  # Tune units in each layer
                activation=hp.Choice("activation", values=["relu", "tanh"]),  # Activation function
            )
        )
    model.add(layers.Dense(1))  # Output layer

    # Compile the model with tunable learning rate
    model.compile(
        optimizer=Adam(learning_rate=hp.Float("learning_rate", 1e-4, 1e-2, sampling="log")),
        loss=rmsre_loss,
    )
    return model

# Initialize the tuner
tuner = kt.Hyperband(
    build_model,
    objective="val_loss",  # Minimize validation loss
    max_epochs=50,
    factor=3,
    directory="hyper_tuning",
    project_name="rmsre_ann_tuning",
)

# Perform the search
tuner.search(X_train, y_train, validation_split=0.2, epochs=50, batch_size=32)

# Get the best hyperparameters and model
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"Best hyperparameters: {best_hps.values}")

# Build and train the best model
best_model = tuner.hypermodel.build(best_hps)
history = best_model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32)

# Evaluate on the test set
test_loss = best_model.evaluate(X_test, y_test)
print(f"Test Loss (RMSRE): {test_loss}")

# Predictions
predictions = best_model.predict(X_test)
print(f'Predictions: {predictions[:5]}')
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import load_model
from tensorflow.keras.applications import ResNet50
import ast

# Load a pre-trained ResNet50 model for feature extraction
base_model = ResNet50(weights='imagenet', include_top=False, pooling='avg')

def load_and_merge_test_data(cgm_test, demo_viome_test, img_test, label_test):
    """Load and merge test data."""
    try:
        # Merge CGM and Image data
        temp_merge = pd.merge(img_test, cgm_test, on=['Subject ID', 'Day'], how='inner')

        # Merge Label data
        temp_merge1 = pd.merge(temp_merge, label_test, on=['Subject ID', 'Day'], how='inner')

        # Check if 'Viome' exists to compute statistics
        if 'Viome' in demo_viome_test.columns:
            viome_means, viome_max, viome_min = [], [], []
            for viome_data in demo_viome_test['Viome']:
                viome_list = [float(x) for x in viome_data.split(',')]
                viome_means.append(np.mean(viome_list))
                viome_max.append(np.max(viome_list))
                viome_min.append(np.min(viome_list))
            demo_viome_test['Viome_Mean'] = viome_means
            demo_viome_test['Viome_Max'] = viome_max
            demo_viome_test['Viome_Min'] = viome_min
            demo_viome_test.drop(columns=['Viome'], inplace=True)
        else:
            print("Warning: 'Viome' column not found. Check the input data!")

        # Merge with Demographics
        merged_test_df = pd.merge(temp_merge1, demo_viome_test, on='Subject ID', how='inner')

        # Ensure the merged dataframe is not empty
        if merged_test_df.empty:
            print("Error: Merged test dataframe is empty. Check your inputs.")
            return None

        return merged_test_df

    except Exception as e:
        print(f"Error during merging: {e}")
        return None

def extract_features_from_image(image_array):
    """Extract features from an image using a pre-trained model."""
    try:
        image_array = np.array(image_array)
        if len(image_array.shape) == 3:  # Ensure it's a single image
            image_array = np.expand_dims(image_array, axis=0)  # Add batch dimension
        features = base_model.predict(image_array, verbose=0)
        return features.flatten()  # Flatten the feature vector
    except Exception as e:
        print(f"Error extracting features: {e}")
        return np.zeros((2048,))  # Return a zero vector if there's an error

def process_image_data(merged_df):
    """Process image data and extract features."""
    merged_df['Image_Before_Breakfast_Features'] = merged_df['Image Before Breakfast'].apply(
        lambda x: extract_features_from_image(ast.literal_eval(x))
    )
    merged_df['Image_Before_Lunch_Features'] = merged_df['Image Before Lunch'].apply(
        lambda x: extract_features_from_image(ast.literal_eval(x))
    )

    # Drop the raw image columns
    merged_df = merged_df.drop(columns=['Image Before Breakfast', 'Image Before Lunch'])

    return merged_df

def process_cgm_data(merged_df):
    """Process CGM Data and extract statistical features."""
    def extract_cgm_stats(cgm_data):
        try:
            cgm_list = ast.literal_eval(cgm_data)  # Convert string to list
            glucose_values = [entry[1] for entry in cgm_list]
            return {
                'Mean_Glucose': np.mean(glucose_values),
                'Max_Glucose': np.max(glucose_values),
                'Min_Glucose': np.min(glucose_values),
                'Std_Glucose': np.std(glucose_values)
            }
        except Exception as e:
            print(f"Error processing CGM Data: {e}")
            return {'Mean_Glucose': 0, 'Max_Glucose': 0, 'Min_Glucose': 0, 'Std_Glucose': 0}

    cgm_features = merged_df['CGM Data'].apply(extract_cgm_stats)
    merged_df['Mean_Glucose'] = cgm_features.apply(lambda x: x['Mean_Glucose'])
    merged_df['Max_Glucose'] = cgm_features.apply(lambda x: x['Max_Glucose'])
    merged_df['Min_Glucose'] = cgm_features.apply(lambda x: x['Min_Glucose'])
    merged_df['Std_Glucose'] = cgm_features.apply(lambda x: x['Std_Glucose'])

    # Drop the original CGM Data column
    merged_df = merged_df.drop(columns=['CGM Data'])

    return merged_df

def prepare_input_for_model(processed_df, expected_input_dim):
    """Prepare input data for the model."""
    # Extract and flatten image features
    image_features = processed_df.apply(
        lambda row: np.concatenate([row['Image_Before_Breakfast_Features'], row['Image_Before_Lunch_Features']]),
        axis=1
    )

    # Drop image feature columns and keep only numerical columns used in training
    non_image_features = processed_df.drop(columns=['Image_Before_Breakfast_Features', 'Image_Before_Lunch_Features'])
    non_image_features = non_image_features.apply(pd.to_numeric, errors='coerce').fillna(0)

    # Convert all data to float32
    image_features = np.stack(image_features).astype('float32')
    non_image_features = non_image_features.values.astype('float32')

    # Concatenate non-image and image features
    input_data = np.hstack([non_image_features, image_features])

    # Match the expected input shape
    if input_data.shape[1] > expected_input_dim:
        input_data = input_data[:, :expected_input_dim]  # Trim excess features
    elif input_data.shape[1] < expected_input_dim:
        padding = np.zeros((input_data.shape[0], expected_input_dim - input_data.shape[1]))
        input_data = np.hstack([input_data, padding])  # Pad with zeros

    return input_data

def preprocess_test_data(merged_df):
    """Preprocess test data to match training pipeline."""
    if merged_df is None or merged_df.empty:
        raise ValueError("Error: The merged test dataframe is empty or invalid.")

    # Convert date and time columns safely
    merged_df['Breakfast_Time'] = pd.to_datetime(merged_df['Breakfast Time'], errors='coerce')
    merged_df['Lunch_Time'] = pd.to_datetime(merged_df['Lunch Time'], errors='coerce')

    # Fill missing or invalid times with default values
    merged_df['Breakfast_Time'].fillna(pd.Timestamp("07:00:00"), inplace=True)  # Default to 7 AM
    merged_df['Lunch_Time'].fillna(pd.Timestamp("12:00:00"), inplace=True)      # Default to 12 PM

    # Add time-based features
    merged_df['Hour_Breakfast'] = merged_df['Breakfast_Time'].dt.hour
    merged_df['Time_Between_Breakfast_Lunch'] = (
        merged_df['Lunch_Time'] - merged_df['Breakfast_Time']
    ).dt.total_seconds()

    # Drop unnecessary columns
    merged_df = merged_df.drop(columns=['Breakfast Time', 'Lunch Time', 'Breakfast_Time', 'Lunch_Time'])

    # Process CGM data
    merged_df = process_cgm_data(merged_df)

    # Process image data
    merged_df = process_image_data(merged_df)

    # Scale continuous features
    continuous_columns = [
        'Subject ID', 'Day', 'Breakfast Calories', 'Breakfast Carbs',
        'Breakfast Fat', 'Breakfast Protein', 'Age', 'Viome_Mean',
        'Viome_Max', 'Viome_Min', 'Hour_Breakfast', 'Time_Between_Breakfast_Lunch',
        'Mean_Glucose', 'Max_Glucose', 'Min_Glucose', 'Std_Glucose'
    ]
    scaler = StandardScaler()
    merged_df[continuous_columns] = scaler.fit_transform(merged_df[continuous_columns])

    return merged_df


# Main workflow
merged_test_df = load_and_merge_test_data(cgm_test, demo_viome_test, img_test, label_test)

if merged_test_df is not None:
    processed_test_df = preprocess_test_data(merged_test_df)

    print("processed_test_df shape:", processed_test_df.shape)

    # Prepare input for model
    expected_input_dim = best_model.input_shape[1]  # Get input feature dimension
    input_data = prepare_input_for_model(processed_test_df, expected_input_dim)

    print("input_data shape:", input_data.shape)

    # Make predictions
    predictions = best_model.predict(input_data)
    predictions = predictions.reshape(-1, 1)  # Ensure it's 2D for inverse_transform

    # Rescale predictions
    predictions_rescaled = scaler_y.inverse_transform(predictions).flatten()

    # Ensure predictions match the number of rows
    if len(predictions_rescaled) != len(processed_test_df):
        raise ValueError(f"Length mismatch: predictions_rescaled={len(predictions_rescaled)}, test_data={len(processed_test_df)}")

    # Create the submission DataFrame
    submission_df = pd.DataFrame({
        'row_id': range(len(predictions_rescaled)),
        'label': predictions_rescaled
    })

    file_path = base_url + "submission.csv"

    # Save the predictions to the specified folder
    submission_df.to_csv(file_path, index=False)
    print(f"Submission file saved at: {file_path}")

else:
    print("Error: Test data merging failed.")
```
